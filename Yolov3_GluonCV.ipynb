```python
from gluoncv import model_zoo, data, utils
from matplotlib import pyplot as plt
import os
import random

net = model_zoo.get_model('yolo3_darknet53_coco', pretrained=True) #downloading pretrained model
```

```python
img_dir = "/content/Real275Data/" #upload your test images to content first
img_names = os.listdir(img_dir)

test_img = random.choice(img_names)
test_img = "0002_color.png"
x, img = data.transforms.presets.yolo.load_test(img_dir+test_img, short=512)

class_IDs, scores, bounding_boxs = net(x)

ax = utils.viz.plot_bbox(img, bounding_boxs[0], scores[0],
                          class_IDs[0], class_names=net.classes)
plt.show()
```
